```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.cvtColor(cv2.imread('rouhani.jpg'), cv2.COLOR_RGB2HSV)
plt.imshow(image)
```

```python
bg = cv2.resize(cv2.imread('hell-park.jpg'),(image.shape[1],image.shape[0]))
plt.imshow(bg)
```

```python
up = np.array([70, 255, 255])
low = np.array([50, 50, 50])
```

```python
mask = cv2.inRange(image, low, up)
res = cv2.bitwise_and(image, image, mask=mask)
plt.imshow(mask)
```

```python
result = image - res
result = cv2.cvtColor(result, cv2.COLOR_HSV2RGB)

result = np.where(result == 0, bg , result)
```

```python
result = cv2.cvtColor(result, cv2.COLOR_RGB2BGR)
plt.imshow(result)
```
